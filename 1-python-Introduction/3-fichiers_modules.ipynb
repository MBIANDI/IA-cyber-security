## 1. Généralités sur les fichiers

### Modes d'ouverture

| Mode | Signification |
|------|---------------|
| `"r"` | Lecture (read) - défaut |
| `"w"` | Écriture (write) - écrase le fichier |
| `"a"` | Ajout (append) - ajoute à la fin |
| `"x"` | Création exclusive |
| `"r+"` | Lecture et écriture |

```python
import os

# La bonne pratique : with open()
with open("exemple.txt", "w", encoding="utf-8") as fichier:
    fichier.write("Hello World!")

# Lecture
with open("exemple.txt", "r", encoding="utf-8") as fichier:
    contenu = fichier.read()
    print(contenu)

# Nettoyage
os.remove("exemple.txt")
```

## 2. Fichiers TXT

```python
import os

# Création d'un fichier de logs
logs = [
    "2024-01-15 10:00:00 - INFO: Démarrage",
    "2024-01-15 10:05:23 - WARNING: Tentative échouée",
    "2024-01-15 10:10:15 - ERROR: Connexion refusée",
    "2024-01-15 10:15:42 - INFO: Arrêt"
]

with open("journal.log", "w", encoding="utf-8") as f:
    for ligne in logs:
        f.write(ligne + "\n")

# Lecture ligne par ligne
print("=== Lecture du journal ===")
with open("journal.log", "r", encoding="utf-8") as f:
    for ligne in f:
        print(ligne.strip())

# Analyse simple
erreurs = 0
with open("journal.log", "r", encoding="utf-8") as f:
    for ligne in f:
        if "ERROR" in ligne:
            erreurs += 1

print(f"\nNombre d'erreurs : {erreurs}")

# Nettoyage
os.remove("journal.log")
```

## 3. Fichiers CSV

```python
import csv
import os

# Écriture d'un fichier CSV
utilisateurs = [
    ["nom", "email", "role"],
    ["Alice", "alice@email.com", "admin"],
    ["Bob", "bob@email.com", "user"],
    ["Charlie", "charlie@email.com", "user"]
]

with open("utilisateurs.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerows(utilisateurs)

# Lecture avec DictReader (recommandée)
with open("utilisateurs.csv", "r", encoding="utf-8") as f:
    reader = csv.DictReader(f)
    for ligne in reader:
        print(f"Nom: {ligne['nom']}, Email: {ligne['email']}")

# Nettoyage
os.remove("utilisateurs.csv")
```

```python
import csv
import os

# Exemple cybersécurité : IOC (Indicateurs de compromission)
iocs = [
    {"type": "IP", "valeur": "192.168.1.105", "risque": "Élevé", "description": "Brute force"},
    {"type": "Domaine", "valeur": "malware-site.com", "risque": "Critique", "description": "C2 connu"},
    {"type": "Hash", "valeur": "5d41402abc4b2a76b9719d911017c592", "risque": "Moyen", "description": "Fichier suspect"}
]

# Écriture
with open("iocs.csv", "w", newline="", encoding="utf-8") as f:
    champs = ["type", "valeur", "risque", "description"]
    writer = csv.DictWriter(f, fieldnames=champs)
    writer.writeheader()
    writer.writerows(iocs)

# Lecture et recherche
def rechercher_ioc(valeur_recherchee):
    with open("iocs.csv", "r", encoding="utf-8") as f:
        reader = csv.DictReader(f)
        for ligne in reader:
            if valeur_recherchee.lower() in ligne["valeur"].lower():
                print(f"Trouvé : {ligne}")

rechercher_ioc("192.168")

os.remove("iocs.csv")
```

## 4. Fichiers JSON

```python
import json
import os

# Configuration d'un pare-feu
config = {
    "mode": "strict",
    "ip_bannies": ["192.168.1.100", "10.0.0.50"],
    "ports_ouverts": [80, 443, 22],
    "log_actif": True,
    "seuil_alertes": 10
}

# Écriture JSON
with open("config.json", "w", encoding="utf-8") as f:
    json.dump(config, f, indent=4, ensure_ascii=False)

# Lecture JSON
with open("config.json", "r", encoding="utf-8") as f:
    config_chargee = json.load(f)
    print(f"Mode : {config_chargee['mode']}")
    print(f"IP bannies : {config_chargee['ip_bannies']}")

# Modification et sauvegarde
config_chargee["ip_bannies"].append("172.16.0.1")

with open("config.json", "w", encoding="utf-8") as f:
    json.dump(config_chargee, f, indent=4)

os.remove("config.json")
```
