## 1. conditions (if, elif, else)

### 1.1 La structure de base : `if`

Une condition permet d'exécuter un bloc de code **si** une condition est vraie.

```python
# Exemple simple
age = 18

if age >= 18:
    print("Vous êtes majeur")

print("Ce message s'affiche toujours")
```

### 1.2 Les opérateurs de comparaison

| Opérateur | Signification | Exemple |
|-----------|---------------|---------|
| `==` | Égal à | `age == 18` |
| `!=` | Différent de | `statut != "admin"` |
| `>` | Supérieur | `paquets > 100` |
| `<` | Inférieur | `tentatives < 3` |
| `>=` | Supérieur ou égal | `niveau >= 5` |
| `<=` | Inférieur ou égal | `port <= 1024` |

```python
# Exercice : Testez les opérateurs
tentatives = 2
SEUIL_MAX = 3

if tentatives >= SEUIL_MAX:
    print("Compte bloqué")
else:
    print(f"Il reste {SEUIL_MAX - tentatives} tentatives")
```

### 1.3 Les opérateurs logiques : `and`, `or`, `not`

```python
# Exemple avec and (les deux conditions doivent être vraies)
ip_valide = True
port_ouvert = True

if ip_valide and port_ouvert:
    print("Connexion autorisée")

# Exemple avec or (au moins une condition vraie)
mdp_correct = False
biometrique_ok = True

if mdp_correct or biometrique_ok:
    print("Authentification réussie")

# Exemple avec not (inverse la condition)
est_banni = False

if not est_banni:
    print("Utilisateur non banni, accès autorisé")
```

### 1.4 La structure complète : `if` / `elif` / `else`

```python
# Classification des ports
port = 443

if port <= 1023:
    print(f"Le port {port} est un port bien connu")
elif port <= 49151:
    print(f"Le port {port} est un port enregistré")
else:
    print(f"Le port {port} est un port dynamique/privé")
```

### 1.5 L'opérateur `in` : tester l'appartenance

```python
# Vérifier si une IP est dans une liste noire
ip_bannies = ["192.168.1.100", "10.0.0.50", "172.16.0.1"]
ip_visiteur = "192.168.1.100"

if ip_visiteur in ip_bannies:
    print(f"ALERTE : IP {ip_visiteur} bannie !")
else:
    print(f"IP {ip_visiteur} autorisée")

# Vérifier si une chaîne contient un mot
log = "Tentative de connexion suspecte détectée"

if "suspecte" in log:
    print("Log suspect identifié")
```

### EXEMPLE CYBERSÉCURITÉ : Système d'authentification

```python
# Simulation d'un système d'authentification
utilisateurs = {
    "admin": "mdp123",
    "alice": "alice2024",
    "bob": "bob_securise"
}

tentatives_echouees = 0
SEUIL_MAX = 3

print("=== SYSTÈME D'AUTHENTIFICATION ===")
login = input("Nom d'utilisateur : ")
mot_de_passe = input("Mot de passe : ")

if tentatives_echouees >= SEUIL_MAX:
    print("[CRITIQUE] Compte bloqué !")
elif login not in utilisateurs:
    print("Nom d'utilisateur inconnu")
    tentatives_echouees += 1
elif utilisateurs[login] != mot_de_passe:
    print("Mot de passe incorrect")
    tentatives_echouees += 1
else:
    print(f"Authentification réussie ! Bienvenue {login}")
    tentatives_echouees = 0
```


## 2. Les boucles

### 2.1 La boucle `for` : parcourir une séquence

La boucle `for` est idéale quand vous connaissez à l'avance le nombre d'itérations.

```python
# Parcourir une liste d'IPs
ips_suspectes = ["10.0.0.1", "192.168.1.105", "172.16.0.89"]

for ip in ips_suspectes:
    print(f"Vérification de l'IP {ip}...")
```

```python
# Utiliser range() pour générer des nombres
print("range(5) : de 0 à 4")
for i in range(5):
    print(f"  i = {i}")

print("\nrange(20, 25) : de 20 à 24")
for port in range(20, 25):
    print(f"  Scan du port {port}")

print("\nrange(80, 8081, 100) : de 80 à 8080 par pas de 100")
for port in range(80, 8081, 100):
    print(f"  Test du port {port}")
```

```python
# enumerate() : obtenir l'index et la valeur
logs = ["Connexion réussie", "Échec login", "Paquet suspect", "Déconnexion"]

for index, log in enumerate(logs):
    print(f"Ligne {index + 1}: {log}")
```

### 2.2 La boucle `while` : répéter tant qu'une condition est vraie

```python
# Simulation d'une attaque par force brute
MOT_DE_PASSE_SECRET = "1234"
tentative = 0
trouve = False

print("=== SIMULATION DE FORCE BRUTE ===")

while tentative <= 9999 and not trouve:
    code_essai = str(tentative).zfill(4)  # Format 0000, 0001, etc.
    if code_essai == MOT_DE_PASSE_SECRET:
        print(f"Code trouvé : {code_essai} après {tentative + 1} essais")
        trouve = True
    else:
        tentative += 1

if not trouve:
    print("Code non trouvé")
```

### 2.3 Contrôle des boucles : `break`, `continue`, `pass`

```python
# break : sort immédiatement de la boucle
print("=== break : arrêt à la première IP dangereuse ===")
ips = ["192.168.1.1", "10.0.0.5", "127.0.0.1", "8.8.8.8"]
ip_dangereuse = "127.0.0.1"

for ip in ips:
    if ip == ip_dangereuse:
        print(f"ALERTE ! IP dangereuse détectée : {ip}")
        break
    print(f"IP {ip} : OK")

# continue : passe à l'itération suivante
print("\n=== continue : ignore certains ports ===")
ports = [22, 80, 443, 3389, 8080, 9999]

for port in ports:
    if port == 3389:  # RDP est souvent bloqué
        print("Port RDP ignoré (politique de sécurité)")
        continue
    print(f"Scan du port {port}")

# pass : ne fait rien (placeholder)
print("\n=== pass : fonction à écrire plus tard ===")
def fonction_a_completer():
    pass  # Pas d'erreur de syntaxe
```

### 🛡️ EXEMPLE CYBERSÉCURITÉ : Détection d'intrusion (IDS simple)

```python
# IDS Simple - Détection de comportements suspects

# Configuration
SEUIL_PORTS_SCANNES = 5
SEUIL_TENTATIVES_FAIL = 3
IP_SURVEILLEES = ["192.168.1.100", "192.168.1.101", "192.168.1.102"]

# Historique des événements
historique = {
    "192.168.1.100": {"ports_scannes": [], "tentatives_echouees": 0},
    "192.168.1.101": {"ports_scannes": [], "tentatives_echouees": 0},
    "192.168.1.102": {"ports_scannes": [], "tentatives_echouees": 0}
}

print("=== SYSTÈME DE DÉTECTION D'INTRUSION ===\n")

for ip, data in historique.items():
    # Simulation de surveillance
    print(f"Surveillance de {ip}")
    
    # Simulation : on ajoute des événements aléatoires
    import random
    nb_evenements = random.randint(1, 8)
    
    for _ in range(nb_evenements):
        type_evenement = random.choice(["connexion", "port_scan"])
        
        if type_evenement == "connexion":
            reussi = random.choice([True, False])
            if not reussi:
                data["tentatives_echouees"] += 1
                if data["tentatives_echouees"] >= SEUIL_TENTATIVES_FAIL:
                    print(f"  *** ALERTE ! Brute force depuis {ip} ***")
        
        else:  # port_scan
            port = random.randint(1, 1000)
            data["ports_scannes"].append(port)
            if len(data["ports_scannes"]) >= SEUIL_PORTS_SCANNES:
                print(f"  *** ALERTE ! Scan réseau depuis {ip} ***")

# Rapport final
print("\n=== RAPPORT FINAL ===")
for ip, data in historique.items():
    print(f"\nIP: {ip}")
    print(f"  Tentatives échouées: {data['tentatives_echouees']}")
    print(f"  Ports scannés: {len(data['ports_scannes'])}")
```

### Exercice 2 : Analyseur de logs

Analysez cette liste de logs et comptez les occurrences de chaque niveau (INFO, WARNING, ERROR).

```python
logs = [
    "INFO: Démarrage du service",
    "ERROR: Connexion à la base échouée",
    "WARNING: Tentative de brute force",
    "INFO: Utilisateur admin connecté",
    "ERROR: Timeout socket",
    "WARNING: Paquet malformé détecté",
    "INFO: Sauvegarde terminée",
    "ERROR: Disk full"
]

# Initialisation des compteurs
info_count = 0
warning_count = 0
error_count = 0

# Parcours des logs
for log in logs:
    if "INFO" in log:
        info_count += 1
    elif "WARNING" in log:
        warning_count += 1
    elif "ERROR" in log:
        error_count += 1

# Affichage des résultats
print(f"INFO: {info_count}")
print(f"WARNING: {warning_count}")
print(f"ERROR: {error_count}")
```

### Exercice 3 : Compteur de tentatives

Créez un programme qui demande un mot de passe et donne 3 tentatives. Si l'utilisateur échoue 3 fois, affichez "Compte bloqué".

```python
# Votre code ici
MOT_DE_PASSE = "python123"
tentatives = 0
MAX_TENTATIVES = 3

while tentatives < MAX_TENTATIVES:
    saisie = input("Entrez le mot de passe : ")
    if saisie == MOT_DE_PASSE:
        print("Accès autorisé")
        break
    else:
        tentatives += 1
        print(f"Mot de passe incorrect. Il vous reste {MAX_TENTATIVES - tentatives} essais")

if tentatives == MAX_TENTATIVES:
    print("Compte bloqué après 3 échecs")
```

## Résumé

| Concept | Syntaxe | Utilité |
|---------|---------|---------|
| Condition | `if condition:` | Exécution conditionnelle |
| Alternative | `elif condition:` | Plusieurs branches |
| Cas par défaut | `else:` | Si aucune condition vraie |
| Boucle for | `for x in sequence:` | Parcours d'une séquence |
| Boucle while | `while condition:` | Répétition conditionnelle |
| Sortie anticipée | `break` | Sortie immédiate |
| Saut d'itération | `continue` | Passe à l'itération suivante |
| Placeholder | `pass` | Ne rien faire |
